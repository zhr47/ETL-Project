```python
csv_path = "Police_Incidents_2015.csv"
police_data = pd.read_csv(csv_path)
```

```python
params = {"key":gkey}

for index, row in police_data.iterrows():
    lat = row['Lat']
    lng = row ['Long']
    
    params["location"] = f"{lat},{lng}"
    base_url = "https://maps.googleapis.com/maps/api/geocode/json?"
    zip_lat_long = f"{base_url}latlng={lat},{lng}&key={gkey}"
    zip_code = requests.get(zip_lat_long)
    zip_code = zip_code.json()
    try:
         police_data.loc[index,"Zip Code"] = zip_code['results'][0]['address_components'][7]['long_name']
    except (KeyError, IndexError):
         print("Missing field/result... skipping.")
```

```python
police_data = police_data.dropna()
```
